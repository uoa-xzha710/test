Data Loading

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Initialize Spark session
spark = SparkSession.builder.appName("agrofood_co2_emission_analysis").getOrCreate()

# Load the data
data = spark.read.csv("final_agrofood_co2_emission.csv", header=True, inferSchema=True)
```

Data Preprocessing

```python
from pyspark.ml.feature import Bucketizer
from pyspark.sql.functions import when, col

# If the Temp_Label column exists, drop it first.
if "Temp_Label" in data.columns:
    data = data.drop("Temp_Label")

# Use Bucketizer for data transformation.
bucketizer = Bucketizer(splits=[-float("inf"), 15, 30, float("inf")], inputCol="Average Temperature °C", outputCol="Temp_Label")
data = bucketizer.transform(data)

# Map the bucket numbers to actual labels.
data = data.withColumn("Temp_Label", col("Temp_Label").cast("int"))
replace_dict = {0: 'Low', 1: 'Medium', 2: 'High'}
for k, v in replace_dict.items():
    data = data.withColumn("Temp_Label", when(col("Temp_Label") == k, v).otherwise(col("Temp_Label")))
```

Data Split

```python
# Split the dataset
train, test = data.randomSplit([0.8, 0.2], seed=42)
```

Regression Modeling

```python
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.feature import VectorAssembler

# Prepare the features
input_columns = [col for col in train.columns if col not in ['Area', 'Year', 'Average Temperature °C', 'Temp_Label']]
assembler = VectorAssembler(inputCols=input_columns, outputCol="features")
train = assembler.transform(train)
test = assembler.transform(test)

# Regression
regressor = RandomForestRegressor(labelCol="Average Temperature °C")
model_reg = regressor.fit(train)
predictions_reg = model_reg.transform(test)
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import StringType
from pyspark.ml.feature import VectorAssembler, StringIndexer
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml import Pipeline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming the data path is the same as before
final_data_path = "final_agrofood_co2_emission.csv"
final_data = spark.read.csv(final_data_path, header=True, inferSchema=True)

# Plot distribution of 'Average Temperature °C'
final_data.select('Average Temperature °C').toPandas().hist(bins=30, edgecolor='k')
plt.title('Distribution of Average Temperature °C')
plt.xlabel('Average Temperature °C')
plt.ylabel('Count')
plt.show()
```

```python
# Create classification labels based on 'Average Temperature °C'
def classify_temp(temp):
    if temp <= 0.5:
        return 'Low'
    elif temp <= 1.5:
        return 'Medium'
    else:
        return 'High'

classify_temp_udf = F.udf(classify_temp, StringType())
data_with_labels = final_data.withColumn('Temp_Label', classify_temp_udf('Average Temperature °C'))
```

```python
input_columns = [col for col in data_with_labels.columns if col not in ['Area', 'Year', 'Average Temperature °C', 'Temp_Label', 'Yearly Average Temperature °C']]
assembler = VectorAssembler(inputCols=input_columns, outputCol='features')
indexer = StringIndexer(inputCol="Temp_Label", outputCol="label")
```

```python
# Train the Random Forest Classifier
rf = RandomForestClassifier(numTrees=50, maxDepth=10, minInstancesPerNode=5, maxBins=int(final_data.count()**(0.5)), seed=42)
pipeline = Pipeline(stages=[assembler, indexer, rf])
model = pipeline.fit(data_with_labels)

```

```python
predictions = model.transform(data_with_labels)
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Classification Accuracy: {accuracy*100:.2f}%")
```

```python
# Get and plot feature importances
importances = model.stages[-1].featureImportances.toArray()
plt.figure(figsize=(12, 6))
indices = np.argsort(importances)[::-1]
plt.title("Feature Importances from Random Forest Classifier")
plt.bar(range(len(importances)), importances[indices], align="center")
plt.xticks(range(len(importances)), np.array(input_columns)[indices], rotation=45)
plt.tight_layout()
plt.show()
```

```python
from pyspark.sql.functions import col, desc

# Convert numpy array data type
importances = [float(imp) for imp in model.stages[-1].featureImportances.toArray()]

# Create a DataFrame with feature names and their importances
feature_importance_df = spark.createDataFrame(list(zip(input_columns, importances)), ["Feature", "Importance"])

# Sort the DataFrame by importance and display the result
feature_importance_sorted = feature_importance_df.orderBy(desc("Importance"))
feature_importance_sorted.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml import Pipeline

# Load data
spark = SparkSession.builder.appName("Agrofood_CO2_Emission").getOrCreate()
data = spark.read.csv("final_agrofood_co2_emission.csv", header=True, inferSchema=True)

# Define the feature columns
input_columns = [col_name for col_name in data.columns if col_name not in ['Area', 'Year', 'Average Temperature °C', 'Temp_Label', 'Yearly Average Temperature °C']]

# Assemble the feature columns into a feature vector
assembler = VectorAssembler(inputCols=input_columns, outputCol="features")

# Split the data for regression task
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)

# Train the Random Forest Regressor, setting the label column
rf_regressor = RandomForestRegressor(labelCol="Average Temperature °C", numTrees=50, maxDepth=10, minInstancesPerNode=5, seed=42)
pipeline_regressor = Pipeline(stages=[assembler, rf_regressor])
model_regressor = pipeline_regressor.fit(train_data)

# Use the model to predict and compute the MSE
from pyspark.ml.evaluation import RegressionEvaluator

predictions_reg = model_regressor.transform(test_data)
evaluator = RegressionEvaluator(labelCol="Average Temperature °C", predictionCol="prediction", metricName="mse")
mse = evaluator.evaluate(predictions_reg)
print(f"Regression MSE: {mse:.4f}")

# Extract and display feature importances
importances_reg = model_regressor.stages[-1].featureImportances.toArray()
feature_importance_df_reg = spark.createDataFrame(list(zip(input_columns, [float(imp) for imp in importances_reg])), ["Feature", "Importance"])

# Sorting the features based on importance
from pyspark.sql.functions import desc

feature_importance_sorted_reg = feature_importance_df_reg.orderBy(desc("Importance"))

# Display the sorted features
feature_importance_sorted_reg.show()



```
